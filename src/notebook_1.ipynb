```python
import pandas as pd
import numpy as np
```

```python
# Funcion para transformar los numeros de los codigos, en el formato correcto
def generar_codigo(number):
    if number == 'NaN':
        return ''
    else:
        return "00" + str(number)
```

```python
df_clients = pd.read_csv('/home/pydev/workflow/dt_unifed_statistics/data/clients_erp_b2b.csv', delimiter=',', encoding='latin1')
df_clients['ID_TERCERO'] =  df_clients['ID_TERCERO'].astype(str) # add client column
df_clients['COD_FRECUENCIA_VISITA']  = df_clients['COD_FRECUENCIA_VISITA'].fillna(0).astype(int)
df_clients['COD_FRECUENCIA_VISITA'] = df_clients['COD_FRECUENCIA_VISITA'].astype(int).astype(str)
# df_clients['COD_FRECUENCIA_VISITA'] =  df_clients['COD_FRECUENCIA_VISITA'].astype(str) # add client column
# df_clients['COD_FRECUENCIA_VISITA'] = df_clients['COD_FRECUENCIA_VISITA'].fillna(0).astype(int)
df_clients.tail(3)
# df_clients.shape
# df_clients['COD_FRECUENCIA_VISITA'].unique()
# df_clients.info()
```

```python
df_visitas = pd.read_csv('/home/pydev/workflow/dt_unifed_statistics/data/data_ruta_visitas_vendedor.csv')
# df_visitas['ID_TERCERO'] =  df_visitas['ID_TERCERO'].astype(str) # add client column
df_visitas.info()
# df_visitas.head(3)
```

```python
df_combined = df_visitas.merge(df_clients, on="ID_TERCERO", how="left")
df_combined.shape
```

```python
# Copy
new_df = df_combined.copy()
# fill nan
new_df = new_df.fillna('')
# Lista de nombres de las columnas que deseas convertir
new_df['COD_FRECUENCIA_VISITA'] = new_df['COD_FRECUENCIA_VISITA'].apply(generar_codigo)
new_df = new_df.drop_duplicates(subset=['ID_TERCERO'])
# Select columns to analyze
# df_columns = ['ID_TERCERO', 'COD_DIA_VISITA','DES_DIA_VISITA','COD_FRECUENCIA_VISITA','DES_FRECUENCIA_VISITA','CANAL']
# apply the columns
new_df = new_df[['ID_TERCERO', 'COD_DIA_VISITA','DES_DIA_VISITA','COD_FRECUENCIA_VISITA','DES_FRECUENCIA_VISITA']]
new_df.shape
```

```python
# Abrir el archivo con las informacion de las fechas asignadas dado, el dia asignado y la frecuencia
df_calendario_Pimovil_Original = pd.read_csv('/home/pydev/workflow/dt_unifed_statistics/data/Calendario_Pimovil.csv')
# Se escogen solo las columnas que son de interes, solo los codigos el resto es redundante
nombres_Pimovil = ["FECHA","COD_DIA VISITA", "COD SEMANAL", "COD QUINCENAL", "COD MENSUAL"]

# df_calendario_Pimovil = df_calendario_Pimovil_Original[nombres_Pimovil]
df_calendario_Pimovil = df_calendario_Pimovil_Original.iloc[:, [0, 1, 3, 5, 7]]

# Funcion para transformar los numeros de los codigos, en el formato correcto
# def generar_codigo(number):
#   return "00"+str(number)

# Lista de nombres de las columnas que deseas convertir
columnas_a_convertir = ["COD_DIA VISITA", "COD SEMANAL", "COD QUINCENAL", "COD MENSUAL"]

#funcion .applymap para recorrer varias columnas y aplicar la tranformacion
df_calendario_Pimovil = df_calendario_Pimovil_Original.copy()
df_calendario_Pimovil[columnas_a_convertir] = df_calendario_Pimovil[columnas_a_convertir].applymap(generar_codigo)

# Usar el método .melt() para "desapilar" las columnas de códigos
df_calendario_Pimovil_UnaTabla = pd.melt(df_calendario_Pimovil, id_vars=["FECHA", "COD_DIA VISITA"], value_vars=["COD SEMANAL",	"COD QUINCENAL","COD MENSUAL"], var_name="Tipo_Codigo", value_name="Codigo_Frecuencia")

# Cambiar el nombre de la columna utilizando el método .rename()
df_calendario_Pimovil_UnaTabla.rename(columns={'COD_DIA VISITA': 'COD_DIA_VISITA','Codigo_Frecuencia':'COD_FRECUENCIA_VISITA', 'FECHA': 'FECHAS_VISITA', 'Tipo_Codigo':'TIPO_CODIGO'}, inplace=True)

df_calendario_Pimovil_UnaTabla.head(3)
```
